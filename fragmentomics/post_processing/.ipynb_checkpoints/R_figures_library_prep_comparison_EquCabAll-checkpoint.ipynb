```r
require("ggpubr")

plt_list <- list()
motif <- "TA"

for (tmp_sample in c("plasma", "BAL")) {
    plt_list[[tmp_sample]] <- list()
    for (tmp_read in c("R1", "R2")) {
        data <- host_nonMT_IS_meta %>%
                    filter(origin %nin% c("controls", "Other molds")) %>%
                    filter(type != "ds-wcDNA") %>%
                    filter(sample == tmp_sample) %>%
                    mutate(read = substring(side, 0, 2)) %>%
                    filter(abs(TLEN) < 250) %>%
                    mutate(TLEN = abs(TLEN)) %>%
                    mutate(EndX__ = paste0(substr(EndMotif, 0, 1), "__")) %>%
                    mutate(End_X_ = paste0("_", substr(EndMotif, 2, 2), "_")) %>%
                    group_by(EndX__, End_X_, sample_id, sample_type, sample, type, patient, origin, read) %>%
                    summarise(sCount = sum(Count)) %>%
                    group_by(sample_id, sample_type, sample, type, patient, origin, read) %>%
                    mutate(nsCount = log10((sCount / sum(sCount, na.rm = TRUE)) * 16)) %>%
                    filter(type == "ss-cfDNA" & read == tmp_read) %>%
                    mutate(stat_test = ifelse(
                        patient %in% c("A01", "A02", "A03") & sample == "plasma" & origin == "IPA samples" |
                        patient %in% c("A01", "A02", "A04", "A14", "A15") & sample == "BAL" & origin == "IPA samples",
                        yes = "positive",
                        no = ifelse(
                            patient %in% c("A04", "A05") & sample == "plasma" & origin == "IPA samples" |
                            patient %in% c("A03", "A05") & sample == "BAL" & origin == "IPA samples" |
                            origin == "External controls",
                            yes = "negative", no = "unknown"
                        )
                    )) %>%
                    mutate(EndXX_ = gsub(paste0(EndX__, End_X_), pattern = "_", replacement = "")) %>%
                    mutate(facets_col = paste0(origin, "\n", stat_test))
        
        stat_test_data <- data %>% 
            group_by(EndXX_) %>% 
            t_test(nsCount ~ facets_col, paired = FALSE, p.adjust.method="bonferroni") 

        stat_test_data %>% 
            filter(p.adj.signif != "ns")

        plt_list[[tmp_sample]][[tmp_read]] <- data %>%
            ggplot(aes(x = facets_col, y = nsCount, col = stat_test)) +
            geom_boxplot(outliers = FALSE) +
            geom_point(aes(color = stat_test), position = position_jitterdodge(dodge.width = 0.6, jitter.width = 0.2), alpha = 0.4, size = 3) +
            facet_wrap(~ EndXX_, scales = "free") +
            ggtitle(paste(tmp_sample, "ss-cfDNA, ", tmp_read)) +
            theme_bw() +
            theme(axis.text.x = element_text(angle = 0)) #+ 
            #stat_pvalue_manual(stat_test_data, label = "p.adj.signif", y.position = 0.5, label.size = 4, step.increase = 0.05)
    }
}

options(repr.plot.width = 20, repr.plot.height = 20)
plt_list[[1]][[1]]
plt_list[[1]][[2]]
plt_list[[2]][[1]]
plt_list[[2]][[2]]
```
